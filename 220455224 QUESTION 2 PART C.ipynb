Import relevant packages

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

Create dataframes to store probabilities and coefficients

```python
prob_df = pd.DataFrame(columns=["Year", "Probability"])
coeff_df = pd.DataFrame(columns=["Year", "Feature", "Coefficient"])
```

Create a dataframe for carriers.csv

```python
carriers_df = pd.read_csv("carriers.csv")
```

Create tables and barplots for each year then fit logistic regression model

```python
# Loop over each year
for i in range(1990, 2000):
    # Read data for each year
    ontime_c = pd.read_csv(f"{i}.csv")
    
    # Left join with carriers_df to include all unique carrier information
    ontime_cc = pd.merge(ontime_c, carriers_df, left_on="UniqueCarrier", right_on="Code", how="left")
    
    # Prepare features and target variable
    features = ["DayOfWeek", "CRSDepTime", "DepTime", "CRSArrTime", "DepDelay", "Distance", "UniqueCarrier"]
    target = "Diverted"
    
    # Drop rows with missing values in the target or features
    ontime_cc.dropna(subset=features + [target], inplace=True)
    
    # Label encode the "UniqueCarrier" column
    label_encoder = LabelEncoder()
    ontime_cc["UniqueCarrier"] = label_encoder.fit_transform(ontime_cc["UniqueCarrier"])
    
    # Fit logistic regression model
    model = LogisticRegression()
    model.fit(ontime_cc[features], ontime_cc[target])
    
    # Predict probabilities of plane being diverted
    probabilities = model.predict_proba(ontime_cc[features])[:, 1]
    
    # Aggregate and store probabilities
    year_prob = np.mean(probabilities)
    prob_df = prob_df.append({"Year": i, "Probability": year_prob}, ignore_index=True)

    # Get coefficients
    coefficients = model.coef_[0]
    
    # Store coefficients and its corresponding year
    coeff_df = coeff_df.append(pd.DataFrame({"Year": [i]*len(coefficients), "Feature": features, "Coefficient": coefficients}), ignore_index=True)
    
    # Sort coefficients in ascending order 
    coeff_sorted = sorted(zip(features, coefficients), key=lambda x: x[1])
    coeff_sorted_features, coeff_sorted_values = zip(*coeff_sorted)
    
    # Create a barplot for each year
    plt.figure()
    plt.barh(coeff_sorted_features, coeff_sorted_values)
    plt.title(f"Logistic Regression Coefficients for Year {i}")
    plt.xlabel("Coefficient Value")
    plt.ylabel("Features")
    plt.show()
    
    # Print coefficients for each year
    print(f"Coefficients for year {i} in ascending order:\n")
    print(pd.DataFrame({"Feature": coeff_sorted_features, "Coefficient": coeff_sorted_values}))

# Plot a line graph for probability of diversion over 1990 to 1999
plt.plot(prob_df["Year"], prob_df["Probability"])
plt.title("Probability of Diverted US Flights Over Time")
plt.xlabel("Year")
plt.ylabel("Probability")
plt.show()
```
